```python
from tqdm import tqdm
```

```python
file1 = 'CN15k/train.txt'
file2 = 'NL27K/train.txt'
file3 = 'UWN18RR/train2id.tsv'
file4 = 'UFB15K237/train2id.tsv'
```

```python
with open(file1,'r') as f1, open(file2,'r') as f2,open(file3,'r') as f3, open(file4,'r') as f4:
    lines1 = f1.readlines()
    lines2 = f2.readlines()
    lines3 = f3.readlines()
    lines4 = f4.readlines()
    data1,data2,data3,data4 = [],[],[],[]
    
    for line in tqdm(lines1):
        data1.append(eval(line.strip().split('\t')[-1]))
    for line in tqdm(lines2):
        data2.append(eval(line.strip().split('\t')[-1]))
    for line in tqdm(lines3):
        data3.append(eval(line.strip().split('\t')[-1]))
    for line in tqdm(lines4):
        data4.append(eval(line.strip().split('\t')[-1]))
```

```python
import matplotlib.pyplot as plt

# 假设 data1 到 data4 是你的四个数据列表
# 设置子图布局为 2x2
fig, axs = plt.subplots(2, 2, figsize=(12, 8))  # 2 行 2 列
# 设置 x 轴坐标范围从 0 到 1
plt.xlim(0, 1)

# 绘制第一个子图（左上角）
axs[0, 0].hist(data1, bins=20,  alpha=0.5, width=0.04)
axs[0, 0].set_title('CN15K')
axs[0, 0].set_xlabel('Confidence', fontsize=12)
axs[0, 0].set_ylabel('Count', fontsize=12)

# 绘制第二个子图（右上角）
axs[0, 1].hist(data2, bins=20,  alpha=0.5, width=0.04)
axs[0, 1].set_title('NL27K')
axs[0, 1].set_xlabel('Confidence', fontsize=12)
axs[0, 1].set_ylabel('Count', fontsize=12)

# 绘制第三个子图（左下角）
axs[1, 0].hist(data3, bins=20,  alpha=0.5, width=0.03)
axs[1, 0].set_title('UWN18RR')
axs[1, 0].set_xlabel('Confidence', fontsize=12)
axs[1, 0].set_ylabel('Count', fontsize=12)

# 绘制第四个子图（右下角）
axs[1, 1].hist(data4, bins=20,  alpha=0.5, width=0.04)
axs[1, 1].set_title('UFB15K237')
axs[1, 1].set_xlabel('Confidence', fontsize=12)
axs[1, 1].set_ylabel('Count', fontsize=12)

# 调整子图之间的间距
plt.tight_layout()

plt.savefig('histogram.eps', format='eps')
# 显示图形
plt.show()
```
